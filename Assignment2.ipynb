```python
# Replacing nan values using imputting
```

```python
num_columns = ['Pregnancies','Glucose',
               'BloodPressure', 'SkinThickness',
               'Insulin','BMI',
               'DiabetesPedigreeFunction','Age']
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=0, strategy='median')
data[num_columns]=imputer.fit_transform(data[num_columns])
```

```python
data.isna().sum()
```

```python
data.hist(bins=10,figsize=(10,10))
plt.show()
```

```python
corrmatrix = data.corr()
plt.subplots(figsize=(15,10))
sns.heatmap(corrmatrix,annot = True, cmap = 'YlGnBu',linewidth = 0.2,vmax = 0.8,vmin = 0.05)
```

```python
data['Glucose'] = data['Glucose'].fillna(data['Glucose'].mode()[0])
```

```python
data['BloodPressure'] = data['BloodPressure'].fillna(data['BloodPressure'].mode()[0])
```

```python
data['BMI'] = data['BMI'].fillna(data['BMI'].mode()[0])
```

```python
data.isna().sum()
```

```python
y = data['Outcome'] # target variable
x = data[['Pregnancies','Glucose',
               'BloodPressure', 'SkinThickness',
               'Insulin','BMI',
               'DiabetesPedigreeFunction','Age']] # features
```

```python
x = pd.get_dummies(data)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state= 42)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')

plt.title('Confusion matrix', y=1.1)


```

```python
# finding the number of patients with and without diabetes

from sklearn.metrics import classification_report
target_names = ['without diabetes', 'with diabetes']
print(classification_report(y_test, y_pred, target_names=target_names))
```

Predicting a peson has diabetes or not

```python
y_predict = log_reg.predict([[	6.0	,148.0,	72.0,	35.0,	125.0	,33.6,	0.627,	50.0,	1]])
print(y_predict)
if y_predict==1:
    print("Diabetic")
else:
    print("Non Diabetic")
```
